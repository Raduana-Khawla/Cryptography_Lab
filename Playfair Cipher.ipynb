```python
import numpy as np
from collections import OrderedDict
# Input.
plaintext ="rafij"
plaintext =plaintext.replace("j","i")
# key.
key = "yaminj"
key = key.replace("j","i")
# Remove duplicate element by using set.
key = "".join(OrderedDict.fromkeys(key))
```

```python
# Creating key matrix.
alphabet = ""
value = 97
for i in range(26):
    if chr(value) == 'j':
        value += 1
    elif chr(value) in key:
        value += 1
    else:
        alphabet += chr(value)
        value += 1

keyMatrix = np.array([[""] * 5] * 5)
i = 0;
k = 0;
for row in range(5):
    for col in range(5):
        if k < len(key):
            keyMatrix[row][col] = key[k]
            k += 1
        else:
            keyMatrix[row][col] = alphabet[i]
            i += 1

inverseKeyMatrix = np.transpose(keyMatrix)

# Encryption.

chphertext = ""
i = 0
plaintext = list(plaintext)

while (i < len(plaintext) - 1):
    if plaintext[i] == plaintext[i + 1]:
        plaintext.insert(i + 1, "x")
        i += 2
    else:
        i += 2
# This condition is for if the number of the plaintext in odd.
if len(plaintext) % 2 != 0:
    plaintext.append("x")

ciphertext = ""
m = 0
while (m < len(plaintext) - 1):
    first = plaintext[m]
    second = plaintext[m + 1]
    
    checkValue = True
    
    if checkValue == True:
        for i in range(5):
            if first in keyMatrix[i] and second in keyMatrix[i]:
                for find in range(5):
                    if keyMatrix[i][find] == first:
                        j = find
                        break
                for find in range(5):
                    if keyMatrix[i][find] == second:
                        k = find
                        break 
              
                ciphertext += keyMatrix[i][(j+1) % 5]
                ciphertext += keyMatrix[i][(k+1) % 5]
                checkValue = False
            
    if checkValue == True:     
        for i in range(5):
            if first in inverseKeyMatrix[i] and second in inverseKeyMatrix[i]:
                for find in range(5):
                    if inverseKeyMatrix[i][find] == first:
                        j = find
                        break
                    
                for find in range(5):
                    if inverseKeyMatrix[i][find] == second:
                        k = find
                        break 
                    
                ciphertext += inverseKeyMatrix[i][(j+1) % 5]
                ciphertext += inverseKeyMatrix[i][(k+1) % 5]
                checkValue = False
        
        if checkValue == True:
            for i in range(5):
                for j in range(5):
                    if keyMatrix[i][j] == first:
                        row1 = i
                        col1 = j
            for i in range(5):
                for j in range(5):
                    if keyMatrix[i][j] == second:
                        row2 = i
                        col2 = j
            ciphertext += keyMatrix[row1][col2]
            ciphertext += keyMatrix[row2][col1]

    m += 2
print(keyMatrix)
print("The corresponding ciphertext is :",ciphertext)
```
